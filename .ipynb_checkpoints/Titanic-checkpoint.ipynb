## 🧾 Dataset Column Descriptions

Here’s a quick explanation of each column in the Titanic dataset:

- **`PassengerId`**: Unique ID assigned to each passenger.
- **`Survived`**: Survival status → `1` = Survived, `0` = Did not survive.
- **`Pclass`**: Ticket class → `1` = First, `2` = Second, `3` = Third class.
- **`Name`**: Full name of the passenger (includes title like Mr., Miss).
- **`Sex`**: Gender of the passenger → `male` or `female`.
- **`Age`**: Age of the passenger (in years); may contain missing values.
- **`SibSp`**: Number of siblings/spouses aboard the Titanic.
- **`Parch`**: Number of parents/children aboard the Titanic.
- **`Ticket`**: Ticket number (somewhat messy and inconsistent).
- **`Fare`**: Ticket price paid by the passenger.
- **`Cabin`**: Cabin number (many missing values).
- **`Embarked`**: Port of embarkation → `C` = Cherbourg, `Q` = Queenstown, `S` = Southampton.

# **Feature Engineering**

**_Let's dive into the most important part of the data analysis. In this part we try to discover some useful column from the given dataset which help to shape our model more effectively_**

### 🛠️ Feature Engineering: Creating the `isalone` Column

To help the model better understand passenger relationships, we engineered a new feature called **`isalone`** based on the existing columns `SibSp` (siblings/spouses aboard) and `Parch` (parents/children aboard).

- **Logic Used:**
  - If `SibSp + Parch == 0`, the passenger is traveling **alone**.
  - Otherwise, they are traveling **with family**.

- **New Column: `isalone`**
  - `1` → Passenger is alone.
  - `0` → Passenger is not alone (has family onboard).

✅ This feature helps highlight that **passengers traveling alone had a lower survival rate** compared to those with family support.


```python
train['f_size']=train['SibSp']+train['Parch']+1
train['isalone']=0
train.loc[train['f_size']==1,'isalone']=1
```

```python
test['f_size']=test['SibSp']+test['Parch']+1
test['isalone']=0
test.loc[test['f_size']==1,'isalone']=1
```

```python
print(train.groupby('isalone')['Survived'].value_counts())
plt.title("survival rate by Passenger Alone status",fontsize=16,color='blue')
sn.countplot(x='isalone',data=train,hue='Survived',palette='pastel')
plt.show()
```

### 🛠️ Feature Engineering: Age and Fare Binning

To simplify continuous variables and extract meaningful categories, the following features were created:

- **`AgeGroup`**: Derived by binning the `Age` column into different age ranges (e.g., child, teen, adult, senior).  
  Helps capture survival trends across age groups and manage missing or noisy age data.

- **`FareBin`**: Created by dividing the `Fare` column into quartiles or price categories (e.g., low, mid, high).  
  Useful for handling skewed fare distribution and improving model performance.

✅ These features make patterns more interpretable and help the model learn better from grouped data.


```python
train['FareBin'] = pd.qcut(train['Fare'], 4, labels=[0, 1, 2, 3])
print(train['FareBin'].value_counts())
```

```python
test['FareBin'] = pd.qcut(test['Fare'], 4, labels=[0, 1, 2, 3])
```

```python

def age_group(age):
    if age <= 4:
        return 'Baby'
    elif age <= 12:
        return 'Child'
    elif age <= 18:
        return 'Teen'
    elif age <= 25:
        return 'YoungAdult'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

# Apply function
train['AgeGroup'] = train['Age'].apply(age_group)
test['AgeGroup'] = test['Age'].apply(age_group)
```

```python
print(train.groupby('AgeGroup')['Survived'].value_counts())

plt.title("Survival count of the people according to age group",fontsize=16,color='green')
sn.countplot(x='AgeGroup',data=train,hue='Survived',palette='Set2')
plt.show()
```

### 🛠️ Feature Engineering: Title Extraction

Extracted titles (Mr, Miss, Mrs, etc.) from the `Name` column to create a new feature `Title_encoded`.  
Grouped rare titles into common categories to reduce noise and improve model clarity.


```python
train['Title'] = train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test['Title'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
rare_titles=['Col','Mlle','Major','Ms',"Mme",'Don','Lady','Sir','Capt','Countess','Jonkheer','Dona']
test['Title']=test['Title'].replace(rare_titles,"Rare")
train['Title']=train['Title'].replace(rare_titles,"Rare")
print(train['Title'].value_counts())
print(train.groupby('Title')['Survived'].value_counts())
```

```python
plt.title("survived count according to title name of the people",fontsize=16,color='blue')
sn.countplot(x='Title',data=train,hue='Survived',palette='Set2')
plt.show()
```


### 🧠 Step 7: Converting Categorical Variables into Numeric

Most machine learning algorithms cannot directly process categorical string values, so we convert them into numeric format:

1. **Sex** → Mapped `male/female` to `1/0`.
2. **AgeGroup** → Binned ages into 5 groups and mapped to `0/1/2/3/4`.
3. **FareBin** → Divided fares into 4 categories and mapped to `0/1/2/3`.
4. **Embarked** → Mapped embarkation ports `S/C/Q` to `0/1/2`.
5. **IsAlone** → Already numeric (`0` = with family, `1` = alone).
6. **Pclass** → Already numeric (`1`, `2`, `3` class).
7. **Title_encoded** → Extracted from name and mapped to numerical categories.

✅ This ensures all features are in a machine-readable numeric format for modeling.


```python
title_map = {
    'Mr': 1,
    'Miss': 2,
    'Mrs': 3,
    'Master': 4,
    'Dr': 5,
    'Rev': 6,
    'Rare': 7
}
title_map2={
    'C':0,
    'Q':1,
    'S':2
}
```

```python
train['Title_encoded'] = train['Title'].map(title_map)
test['Title_encoded'] = test['Title'].map(title_map)
```

```python
train['embark']=train['Embarked'].map(title_map2).astype(int)
test['embark']=test['Embarked'].map(title_map2).astype(int)
```

```python
le=LabelEncoder()
train['Sex']=le.fit_transform(train['Sex'])
test['Sex']=le.fit_transform(test['Sex'])
```

```python
le_age = LabelEncoder()
train['AgeGroup'] = le_age.fit_transform(train['AgeGroup'])
test['AgeGroup'] = le_age.fit_transform(test['AgeGroup'])
```

```python
print(train.columns)
print(test.columns)
```

**Now let's drop all the unnecessary column which are not important for training the model**

```python
train.drop(['PassengerId','Name','Age','SibSp','Parch','Fare','Embarked','f_size','Title','Ticket','Cabin'],inplace=True,axis=1,errors='ignore')
test.drop(['Name','Age','SibSp','Parch','Fare','Embarked','f_size','Title'],inplace=True,axis=1,errors='ignore')
```

### **Now let's apply the Model**

```python
x=train.drop('Survived',axis=1)
y=train['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=66)
models = [
    LogisticRegression(max_iter=1000),
    LinearSVC(max_iter=10000),
    SVC(kernel='rbf'),
    KNeighborsClassifier(),
    RandomForestClassifier(),
    DecisionTreeClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()
]
model_names = [
    'LogisticRegression',
    'LinearSVM',
    'rbfSVM',
    'KNearestNeighbors',
    'RandomForest',
    'DecisionTree',
    'GradientBoosting',
    'GaussianNB',
    'LinearDiscriminant',
    'QuadraticDiscriminant'
]
accuracy=[]
for clf in models:
    clf.fit(x_train, y_train)
    pred = clf.predict(x_test)
    acc = accuracy_score(y_test, pred)
    accuracy.append(acc)

# 🪞 Compare results
compare = pd.DataFrame({
    'Algorithm': model_names,
    'Accuracy': accuracy
}).sort_values(by='Accuracy', ascending=False).reset_index(drop=True)

print(compare)
```

```python
plt.figure(figsize=(10, 6))
sn.pointplot(data=compare, x='Accuracy', y='Algorithm', color='blue', markers='o', linestyles='-')
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(0.7, 0.9)  # Adjust based on your accuracy range
plt.tight_layout()
plt.show()
```

### 📊 Model Comparison and Accuracy

Multiple classification algorithms were trained and evaluated on the Titanic dataset.  
Below is a comparison of their accuracy scores:

- 🚀 **Gradient Boosting** achieved the highest accuracy of **~85%**
- 🌳 **Decision Tree** and **Random Forest** also performed well with over **83%**
- 🧠 Simpler models like **Logistic Regression** and **Naive Bayes** showed lower accuracy
- 🔍 The graph below visualizes how each model performed on the same dataset

✅ This helped identify the best-performing model to use for final predictions.


### 📌 Evaluating the Best Model — Gradient Boosting Classifier

After comparing various models, the **Gradient Boosting Classifier** was selected as the best performer based on accuracy.  
To evaluate its performance more deeply, we used a **confusion matrix** and **classification report**:

- The **confusion matrix** shows the number of correct and incorrect predictions for each class (survived or not).
- The **classification report** provides detailed metrics:
  - **Precision**: How many predicted positives were actually correct.
  - **Recall**: How many actual positives were correctly predicted.
  - **F1-score**: A balance between precision and recall.
  - **Support**: Number of samples in each class.

✅ These metrics help understand the model’s performance **beyond just accuracy**, especially in handling class imbalances (e.g., more non-survivors than survivors).


```python
best_model = GradientBoostingClassifier()
best_model.fit(x_train, y_train)
y_pred = best_model.predict(x_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sn.heatmap(cm, annot=True, fmt="d", cmap='YlGnBu')
plt.title('Confusion Matrix (Gradient Boosting)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```
